```python
#Filtrando colunas de interesse

flt = (base.League == "WORLD - WORLD CUP") & (base.Season != 2022)
df = base[flt]
df = df[['Season','Home','Away','FT_Odd_H','FT_Odd_D','FT_Odd_A','FT_Goals_H','FT_Goals_A']]

df['CV_Odds'] = df[['FT_Odd_H','FT_Odd_D','FT_Odd_A']].std(ddof=0, axis=1) / df[['FT_Odd_H','FT_Odd_D','FT_Odd_A']].mean(axis=1)

df.reset_index(inplace=True, drop=True)
df.index = df.index.set_names(['Nº'])
df = df.rename(index=lambda x: x + 1)
df
```

```python
copa = base[(base.League == "WORLD - WORLD CUP") & (base.Season == 2022)]
copa = copa[['Season','Home','Away','FT_Odd_H','FT_Odd_D','FT_Odd_A','FT_Goals_H','FT_Goals_A']]

copa['CV_Odds'] = copa[['FT_Odd_H','FT_Odd_D','FT_Odd_A']].std(ddof=0, axis=1) / copa[['FT_Odd_H','FT_Odd_D','FT_Odd_A']].mean(axis=1)
#coeficiente de variação divido por media

copa.reset_index(inplace=True, drop=True)
copa.index = copa.index.set_names(['Nº'])
copa = copa.rename(index=lambda x: x + 1)

copa
```

```python
fltro1 = (copa.FT_Odd_H < copa.FT_Odd_A) & (copa.CV_Odds <= 0.399782)
back_home = copa[fltro1]
back_home
```

```python
fltro1 = (copa.FT_Odd_H > copa.FT_Odd_A) & (copa.CV_Odds >= 0.399782)
back_away = copa[fltro1]
back_away
```
